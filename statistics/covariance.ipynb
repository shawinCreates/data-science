```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('../dataset/test.csv')
dataset.head(5)
```

```python
dataset.info()
```

```python
data_corr = dataset.select_dtypes(['float64','int64']).corr()
data_corr
```

```python

sns.heatmap(data_corr, annot = True)
plt.title('Corelation Matrix Heatmap')
plt.show()
```

```python
data_cov = dataset.select_dtypes(['float64']).cov()
data_cov
```

```python
sns.heatmap(data_cov, annot = True)
plt.show()
```
