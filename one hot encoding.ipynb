```python
dt2 = pd.read_csv("thyroid_cancer_risk_data.csv")
dt2.head()
```

```python
#get dummies method
encoded = dt2[["Family_History","Radiation_Exposure", "Iodine_Deficiency", "Smoking"]]
pd.get_dummies(encoded)
```

```python
dt2['Ethnicity'].unique()
```

```python
#using scikit
encoded = dt2[['Gender','Ethnicity', 'Smoking', 'Obesity', 'Diabetes']]
```

```python
ohe1 = OneHotEncoder(drop='first', sparse_output=False)
arr = ohe1.fit_transform(encoded)
a_columns = ohe1.get_feature_names_out(['Gender','Ethnicity', 'Smoking', 'Obesity', 'Diabetes'])
```

```python
en_dataframe = pd.DataFrame(arr, columns=a_columns)
en_dataframe
```
