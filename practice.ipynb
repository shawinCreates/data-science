```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Coordinates of the point
x = [2]
y = [3]

# Create the plot
plt.scatter(x, y, color='red', label='Point (2, 3)')

# Add labels and grid
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Single Point Plot')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define two points: (x1, y1) and (x2, y2)
x = [2, 56]
y = [55, 48]

# Plot the line
plt.plot(x, y, color='blue', linewidth=2, label='Line from (1,2) to (4,8)')

# Add labels and grid
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np

x = np.linspace(0, 10, 100)
y = 2 * x + 1

plt.plot(x, y, label='y = 2x + 1', color='green')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line: y = 2x + 1')
plt.grid(True)
plt.legend()
plt.show()
```

```python
dataset = pd.read_csv('Train.csv')
plt.bar(dataset['Category'],dataset['Sales_Quantity'])
```

```python
mean_sales = dataset.groupby('Category')['Sales_Quantity'].mean()
mean_sales.plot(kind='bar')
```

```python
mean_sales = dataset.groupby(['Category', 'Brand'])['Sales_Quantity'].mean().unstack()


mean_sales.plot(kind='bar', stacked=True)
plt.title('Mean Sales per Brand in Each Category')
plt.ylabel('Average Sell Quantity')

plt.show()
```

```python
sales_by_category = dataset.groupby('Category')['Sales_Quantity'].sum()

# Plot pie chart
plt.pie(sales_by_category, labels=sales_by_category.index, autopct='%1.1f%%')
plt.axis('equal')
plt.title('Sales Distribution by Category')
plt.show()
```
