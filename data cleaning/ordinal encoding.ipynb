```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.read_csv("../dataset/Students_Grading_Dataset.csv")
df.head()
```

```python
df.isnull().sum()/df.shape[0]*100
```

```python
df['Parent_Education_Level'].fillna(df['Parent_Education_Level'].mode()[0], inplace=True)
```

```python
df.head()
```

```python
df['Parent_Education_Level'].unique()
```

```python
ord_data = [['High School', "Bachelor's", "Master's", 'PhD']]
```

```python
#encode using cyclic line
oe = OrdinalEncoder(categories=ord_data)
encoded = oe.fit_transform(df[['Parent_Education_Level']])
```

```python
df['encoded_parent_education_level'] = encoded
df.head()
```

```python
df['Family_Income_Level'].unique()
```

```python
#using map function
en_family_income = {'Low': 1, 'Medium':2, 'High':3}
```

```python
df['en_family_income'] = df['Family_Income_Level'].map(en_family_income)
df.head()
```
