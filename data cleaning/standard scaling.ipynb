```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
dataset = pd.read_csv('../dataset/laptop_prices.csv')
dataset.head()
```

```python
sns.kdeplot(dataset['Price'], fill=True)
plt.show()
```

```python
dataset.describe()
```

```python
ss = StandardScaler()
ss.fit(dataset[['Price']])
dataset[['Scaled_price']] = ss.transform(dataset[['Price']])
dataset.head()
```

```python
dataset.head()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
plt.title("Before")
sns.kdeplot(dataset['Price'], fill=True)

plt.subplot(1, 2, 2)
plt.title("After")
sns.kdeplot(dataset['Scaled_price'], fill=True)
```
