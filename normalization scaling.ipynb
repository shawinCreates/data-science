```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
```

```python
dataset = pd.read_csv('cereal.csv')
dataset.head()
```

```python
dataset.describe()
```

```python
sns.kdeplot(dataset['potass'])
plt.show()
```

```python
ms = MinMaxScaler()
ms.fit(dataset[['potass']])
dataset[['min_potass']] = ms.transform(dataset[['potass']])
dataset.head()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
plt.title("Before")
sns.kdeplot(dataset['potass'], fill=True)

plt.subplot(1, 2, 2)
plt.title("After")
sns.kdeplot(dataset['min_potass'], fill=True)
```
