```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
org_dataset = pd.read_csv(r'../dataset/iris.csv')
org_dataset.head()
```

```python
unlabeled_dataset = org_dataset.drop(columns='variety')
unlabeled_dataset.head()
```

```python
sns.pairplot(data=unlabeled_dataset)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []

for i in range(2,21):
    kmc = KMeans(n_clusters=i, init='k-means++')
    kmc.fit(unlabeled_dataset)
    wcss.append(kmc.inertia_)
```

```python
plt.figure(figsize=(10,5))
plt.plot([i for i in range(2,21)], wcss, marker='o')
plt.xlabel("No. of Clusters")
plt.ylabel("WCSS value")
plt.xticks([i for i in range(2,21)])
plt.grid(axis="both")
plt.show()
```

```python
km = KMeans(n_clusters=3, init='k-means++')
unlabeled_dataset['predict'] = km.fit_predict(unlabeled_dataset)
unlabeled_dataset.head()
```

```python
sns.pairplot(data=unlabeled_dataset, hue='predict')
plt.show()
```
