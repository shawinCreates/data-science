```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
org_dataset = pd.read_csv(r'../dataset/iris.csv')
org_dataset.head()
```

```python
unlabeled_dataset = org_dataset.drop(columns='variety')
unlabeled_dataset.head()
```

```python
sns.pairplot(data=unlabeled_dataset)
plt.show()
```

```python
import scipy.cluster.hierarchy as sc
```

```python
plt.figure(figsize=(20,10))
sc.dendrogram(sc.linkage(unlabeled_dataset, method='single', metric='euclidean'))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac = AgglomerativeClustering(n_clusters=2, linkage='single')
unlabeled_dataset['predict'] = ac.fit_predict(unlabeled_dataset)
```

```python
unlabeled_dataset.head()
```

```python
sns.pairplot(unlabeled_dataset, hue='predict')
plt.show()
```
