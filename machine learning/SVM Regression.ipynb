```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv(r'../dataset/linear_regression_practice.csv')
dataset.head()
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
sns.scatterplot(x='Study_Hours', y='Exam_Score', data=dataset)
plt.show()
```

```python
x = dataset[['Study_Hours']]
y = dataset['Exam_Score']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
from sklearn.svm import SVR
```

```python
sv1 = SVR(kernel='linear')
sv1.fit(x_train, y_train)
```

```python
sv1.score(x_train, y_train)*100,sv1.score(x_test, y_test)*100
```

```python
sns.scatterplot(x='Study_Hours', y='Exam_Score', data=dataset)
plt.plot(dataset['Study_Hours'], sv1.predict(x), color = 'red')
plt.show()
```

```python
sv2 = SVR(kernel='poly', degree=5)
sv2.fit(x_train, y_train)
```

```python
sv2.score(x_train, y_train)*100, sv2.score(x_test, y_test)*100
```

```python
sns.scatterplot(x='Study_Hours', y='Exam_Score', data=dataset)
plt.plot(dataset['Study_Hours'], sv2.predict(x), color = 'red')
plt.show()
```

```python
sv3 = SVR(kernel='rbf')
sv3.fit(x_train, y_train)
```

```python
sv3.score(x_train, y_train)*100, sv3.score(x_test, y_test)*100
```

```python
sns.scatterplot(x='Study_Hours', y='Exam_Score', data=dataset)
plt.plot(dataset['Study_Hours'], sv3.predict(x), color = 'red')
plt.show()
```

```python
sv4 = SVR(kernel='sigmoid')
sv4.fit(x_train, y_train)
```

```python
sv4.score(x_train, y_train)*100, sv4.score(x_test, y_test)*100
```

```python
sns.scatterplot(x='Study_Hours', y='Exam_Score', data=dataset)
plt.plot(dataset['Study_Hours'], sv4.predict(x), color = 'red')
plt.show()
```
