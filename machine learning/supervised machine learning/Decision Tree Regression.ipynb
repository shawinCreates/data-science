```python
import pandas as pd
```

```python
dataset = pd.read_csv(r'../dataset/cardekho_data.csv')
dataset.head()
```

```python
x = dataset.iloc[:, 3:5]
y = dataset['Selling_Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor, plot_tree
```

```python
for i in range (1,20):
    dt = DecisionTreeRegressor(max_depth=i, criterion='squared_error')
    dt.fit(x_train, y_train)
    print(i, dt.score(x_train,y_train)*100, dt.score(x_test, y_test)*100)
```

```python
dt1 = DecisionTreeRegressor(max_depth=13, criterion='squared_error')
dt1.fit(x_train,y_train)
```

```python
plot_tree(dt1)
```
