```python
import pandas as pd
```

```python
dataset = pd.read_csv(r'../dataset/linear_regression_practice.csv')
dataset.head(10)
```

```python
x = dataset.iloc[:, :-1]
y = dataset['Exam_Score']
```

```python
new_data = dataset.head(10)
```

```python
x_new = new_data.iloc[:, :-1]
y_new = new_data['Exam_Score']
```

```python
from sklearn.model_selection import LeaveOneOut, LeavePOut, KFold, StratifiedKFold
```

```python
lo =  LeaveOneOut()

for train, test in lo.split(x_new, y_new):
    print(train, test)
```

```python
lpo =  LeavePOut(p=4)

for train, test in lpo.split(x_new, y_new):
    print(train, test)
```

```python
kf =  KFold(n_splits=8)

for train, test in kf.split(x_new, y_new):
    print(train, test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(LinearRegression(), x, y, cv=10)
```

```python
cross_val_score(LinearRegression(), x, y, cv=LeavePOut(p=2))
```

```python
cross_val_score(LinearRegression(), x, y, cv=KFold(n_splits=15))
```
