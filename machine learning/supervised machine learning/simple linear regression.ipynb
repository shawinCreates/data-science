```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv('../dataset/exam_score.csv')
dataset.head()
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x='Study_Hours', y='Exam_Score', data=dataset)
plt.show()
```

```python
x = dataset[['Study_Hours']]
y = dataset[['Exam_Score']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
lr = LinearRegression()
lr.fit(x, y)
```

```python
lr.predict([[6.43]])
```

```python
lr.score(x_test,y_test)*100
```

```python
predict_score = lr.predict(x)
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x='Study_Hours', y='Exam_Score', data=dataset)
plt.plot(dataset['Study_Hours'], predict_score, c="red")
plt.legend(['Original Data', 'Predict Line'])
plt.show()
```
