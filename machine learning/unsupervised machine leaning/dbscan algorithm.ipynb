```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons
```

```python
x, y = make_moons(n_samples=250, noise=0.05)
```

```python
df = {"data1":x[:, 0], "data2":x[:, 1], "output":y}
```

```python
dataset = pd.DataFrame(df)
dataset.head()
```

```python
sns.scatterplot(x='data1', y="data2", data=dataset)
plt.show()
```

```python
unlabeled_dataset = dataset.iloc[:, :-1]
unlabeled_dataset.head()
```

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps=0.25, min_samples=5, metric='euclidean')
unlabeled_dataset['predict'] = db.fit_predict(unlabeled_dataset)
```

```python
unlabeled_dataset.head()
```

```python
sns.scatterplot(x='data1', y="data2", data=unlabeled_dataset, hue='predict')
plt.show()
```
