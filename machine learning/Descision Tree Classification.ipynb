```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv(r"../dataset/Social_Network_Ads.csv")
dataset.head()
```

```python
dataset = dataset.drop(columns=['User ID', 'Gender'])
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
sns.scatterplot(data=dataset, x='Age', y="EstimatedSalary", hue="Purchased")
plt.show()
```

```python
x = dataset.iloc[:, :-1]
y = dataset['Purchased']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
sc.fit(x)
x = pd.DataFrame(sc.transform(x))
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
dt.score(x_test,y_test)*100
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plt.figure(figsize=(3,3))
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=dt)
plt.show()
```

```python
plt.figure(figsize=(50,50))
plot_tree(dt)
plt.savefig("classificationDecisionTree.png")
plt.show()
```
